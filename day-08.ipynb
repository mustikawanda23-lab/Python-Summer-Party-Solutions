## Question 3

Between April 1st and June 30th, 2025, what would be the predicted sales lift if a 'pay over time' option were introduced? Assume that 20% of credit card transactions during this period would switch to using the 'pay over time' option. And that for these switched transactions, the order value is expected to increase by 15% based on the average order value of all credit card transactions in that same time period.

```python
# Filter for April 1 - June 30, 2025
fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'])
mask = (fct_transactions['transaction_date'] >= '2025-04-01') & (fct_transactions['transaction_date'] <= '2025-06-30')
df_period = fct_transactions.loc[mask]

# Identify unique customers using credit card
credit_card_customers = df_period[df_period['payment_method'] == 'credit_card']['customer_id'].unique()

# Select 20% of those customers
num_switch_customers = int(len(credit_card_customers) * 0.20)
switch_customers = np.random.choice(credit_card_customers, num_switch_customers, replace=False)

# Apply 15% increase for their transactions
df_period['adjusted_order_value'] = df_period.apply(
    lambda row: row['order_value'] * 1.15 if row['customer_id'] in switch_customers and row['payment_method'] == 'credit_card'
    else row['order_value'],
    axis=1
)

# Calculate original and adjusted sales totals
original_sales = df_period['order_value'].sum()
adjusted_sales = df_period['adjusted_order_value'].sum()

# Calculate predicted sales lift
sales_lift = adjusted_sales - original_sales
sales_lift_pct = (sales_lift / original_sales) * 100

print(f"Original Sales: ${original_sales:,.2f}")
print(f"Adjusted Sales: ${adjusted_sales:,.2f}")
print(f"Predicted Sales Lift: ${sales_lift:,.2f} ({sales_lift_pct:.2f}%)")
```
