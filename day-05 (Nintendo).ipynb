## Question 2

Using the cleaned data, calculate the total pre-sale orders per month for each region and demographic group.

```python
# Ensure date column is datetime type
df['pre_order_date'] = pd.to_datetime(df['pre_order_date'], errors='coerce')

# Drop rows with invalid dates just in case
df = df.dropna(subset=['pre_order_date'])

# Extract year and month
df['order_month'] = df['pre_order_date'].dt.to_period('M')  # e.g., 2024-09

# Group by region, demographic_group, and month → then sum pre_order_quantity
monthly_orders = df.groupby(
    ['region', 'demographic_group', 'order_month']
)['pre_order_quantity'].sum().reset_index()

# Rename for clarity
monthly_orders.rename(columns={'pre_order_quantity': 'total_pre_orders'}, inplace=True)

# Sort
monthly_orders = monthly_orders.sort_values(by=['order_month', 'region', 'demographic_group'])

# Display result
print(monthly_orders)
```

## Question 3

Predict the total pre-sales quantity for each region for September 2024. Assume that growth rate from August to September, is the same as the growth rate from July to August in each region.

```python
# Ensure date column is datetime
df['pre_order_date'] = pd.to_datetime(df['pre_order_date'], errors='coerce')
df = df.dropna(subset=['pre_order_date'])

# Extract year-month period
df['order_month'] = df['pre_order_date'].dt.to_period('M')

# Filter for July and August 2024 only
df_filtered = df[
    df['order_month'].isin([pd.Period('2024-07'), pd.Period('2024-08')])
]

# Total pre-orders per region per month
monthly_totals = df_filtered.groupby(['region', 'order_month'])['pre_order_quantity'].sum().unstack(fill_value=0)
monthly_totals.columns = ['July', 'August']

# Calculate growth rate: (Aug - July) / July
monthly_totals['growth_rate'] = (
    (monthly_totals['August'] - monthly_totals['July']) / monthly_totals['July']
).replace([float('inf'), -float('inf')], 0).fillna(0)

# Predict September = August * (1 + growth_rate)
monthly_totals['September_Predicted'] = (monthly_totals['August'] * (1 + monthly_totals['growth_rate'])).round()

# Reset index to display result nicely
forecast_result = monthly_totals[['July', 'August', 'growth_rate', 'September_Predicted']].reset_index()

# Display result
print("Predicted pre-orders for September 2024 by region:")
print(forecast_result)
```
