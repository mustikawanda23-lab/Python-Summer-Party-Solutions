## Question 2

After dropping duplicates, aggregate the data to find the average user interaction time for each number of search results displayed per page. What are the average interaction times?

```python
# Drop duplicate rows (all columns)
df_cleaned = df.drop_duplicates()

# Group by number of search results displayed
avg_interaction = df_cleaned.groupby('search_results_displayed')['interaction_time'].mean().reset_index()

# Rename for clarity
avg_interaction.rename(columns={'interaction_time': 'avg_interaction_time'}, inplace=True)

# Optional: sort by number of search results
avg_interaction = avg_interaction.sort_values(by='search_results_displayed')

# Show result
print(avg_interaction)
```

## Question 3

Sort the aggregated results from Q2 to determine which number of search results per page has the highest average user interaction time. What is the optimal number of search results per page?

```python
# Clean the dataset
df_cleaned = df.drop_duplicates()

# Calculate average interaction time by search_results_displayed
avg_interaction = df_cleaned.groupby('search_results_displayed')['interaction_time'].mean().reset_index()

# Rename column
avg_interaction.rename(columns={'interaction_time': 'avg_interaction_time'}, inplace=True)

# Sort by avg_interaction_time in descending order to find the maximum
sorted_avg = avg_interaction.sort_values(by='avg_interaction_time', ascending=False)

# Get the top value
optimal_row = sorted_avg.iloc[0]

# Display the full sorted table and the optimal value
print("Sorted average interaction times:")
print(sorted_avg)

print("\nOptimal number of search results per page:")
print(f"Search Results: {optimal_row['search_results_displayed']}, "
      f"Average Interaction Time: {round(optimal_row['avg_interaction_time'], 2)} seconds")
```
