## Question 1

Identify and list all unique customer IDs who have made returns between July 1st 2024 and June 30th 2025. This will help us understand the base set of customers involved in returns during the specified period.

```python
# Normalize and parse the date column (column name has a space: 'order date')
customer_returns['order_date'] = pd.to_datetime(customer_returns['order_date'], errors='coerce')

# Normalize the return flag to boolean (covers various representations)
# Convert to string, strip, lowercase for safe comparisons
flag_str = customer_returns['return_flag'].astype(str).str.strip().str.lower()

# Define values that indicate a return
true_vals = {'1', 'true', 't', 'yes', 'y', 'returned', 'return', 'r'}

customer_returns['_is_return'] = flag_str.isin(true_vals)

# Filter by date range and where _is_return == True
start = pd.Timestamp('2024-07-01')
end = pd.Timestamp('2025-06-30')

mask = (
    (customer_returns['order_date'] >= start) &
    (customer_returns['order_date'] <= end) &
    (customer_returns['_is_return'])
)

returns_period = customer_returns.loc[mask]

# Get unique customer IDs
unique_customers = returns_period['customer_id'].dropna().unique()

# Return as a DataFrame for display/export and show count
unique_customers_df = pd.DataFrame({'customer_id': unique_customers})
print(f"Number of unique customers with returns in period: {len(unique_customers)}")
print(unique_customers_df.head(50))  # show first 50 as a sample

# Full list as Python list
unique_customers_list = unique_customers.tolist()
```

## Question 2

Convert the 'order_date' column to a datetime format and create a MultiIndex with 'customer_id' and 'order_date'. Then, calculate the total number of returns per customer for each month. This will provide insights into monthly return patterns for each customer.

```python
# Parse order_date robustly
try:
    customer_returns['order_date'] = pd.to_datetime(
        customer_returns['order_date'],
        format='%Y-%m-%d',
        errors='raise'
    )
except Exception:
    customer_returns['order_date'] = pd.to_datetime(
        customer_returns['order_date'],
        errors='coerce'
    )

# Report any unparsed dates
n_unparsed = customer_returns['order_date'].isna().sum()
print(f"Unparsed order_date rows after parsing: {n_unparsed}")

# Normalize return flag column name and values
if 'return_flag' in customer_returns.columns:
    flag_col = 'return_flag'
elif 'return flag' in customer_returns.columns:
    flag_col = 'return flag'
else:
    raise KeyError("No 'return_flag' or 'return flag' column found in customer_returns.")

# Normalize to boolean
flag_vals = customer_returns[flag_col].astype(str).str.strip().str.lower()
true_vals = {'1','true','t','yes','y','returned','return','r'}
customer_returns['_is_return'] = flag_vals.isin(true_vals)

# Filter returned orders and valid dates in the target period if needed
# start = pd.Timestamp('2024-07-01'); end = pd.Timestamp('2025-06-30')
# customer_returns = customer_returns[(customer_returns['order_date'] >= start) & (customer_returns['order_date'] <= end)]

returns_only = customer_returns[
    customer_returns['_is_return'] & customer_returns['order_date'].notna()
].copy()

# Create 'year_month' period for grouping
returns_only['year_month'] = returns_only['order_date'].dt.to_period('M')

# Create MultiIndex and compute monthly totals per customer
returns_multi = returns_only.set_index(['customer_id', 'order_date']).sort_index()

monthly_returns = (
    returns_only
    .groupby(['customer_id', 'year_month'])
    .size()
    .rename('total_returns')
)

# Show result
print("Sample monthly returns (customer_id, year_month -> total_returns):")
print(monthly_returns.head(20))
```
