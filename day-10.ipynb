## Question 3

Calculate the basic summary statistics (mean, median, standard deviation) of app ratings for each category to identify variations and performance patterns.

```python
# Remove leading/trailing whitespace in rating
df['rating'] = df['rating'].astype(str).str.strip()

# Replace commas with decimal points
df['rating'] = df['rating'].str.replace(',', '.', regex=False)

# Convert to numeric, coercing non-numeric values to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Group by category and calculate summary statistics
summary_stats = df.groupby('category')['rating'].agg(
    mean_rating='mean',
    median_rating='median',
    std_rating='std'
).reset_index()

# Display the results
print(summary_stats)
```
